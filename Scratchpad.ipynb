```python
import pandas as pd
df = pd.read_csv('nyc-discretionary-funding_2021.csv')
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.replace("-", "_")
df.columns = df.columns.str.lower()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df.head(10)
```

```python
df.query('source == "Local"').legal_name_of_organization.value_counts()
```

```python
new_df = df.groupby(by='source').amount.sum().reset_index().sort_values(by='amount', ascending = False).head(5)
```

```python
df.groupby(by='legal_name_of_organization').amount.sum().reset_index().sort_values(by='amount', ascending = False).head(5)
```

```python
df.query('legal_name_of_organization == "Legal Aid Society"')
```

```python
new_df
```

```python
new_df.to_csv('top_5.csv')
```
