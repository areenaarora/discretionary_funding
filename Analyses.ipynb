```python
import pandas as pd
df_2021 = pd.read_csv('nyc-discretionary-funding_2021.csv')
df_2021.columns = df_2021.columns.str.replace(" ", "_")
df_2021.columns = df_2021.columns.str.replace("-", "_")
df_2021.columns = df_2021.columns.str.lower()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:,.2f}'.format
```

## Top five recepients in 2021

```python
top_5_2021 = df_2021.groupby(by='legal_name_of_organization').amount.sum().reset_index().sort_values(by='amount', ascending = False).head(5)
```

```python
top_5_2021
```

```python
top_5_2021.amount.sum()
```

```python
df_2021.amount.sum()
```

```python
top_5_2021.to_csv('top_5_2021.csv')
```
